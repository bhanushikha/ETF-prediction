## The Intricacies of Financial Time series

Financial time series stand as a sequential tapestry of data points, systematically captured at uniform time junctures. This form of temporal data holds paramount importance across diverse realms, encompassing fields like finance, economics, atmospheric sciences, and engineering. Delving into financial time series unravels an ensemble of trends, trajectories, and tell-tale signs that underpin decision matrices and predictive forays. Nonetheless, when navigating the vast expanse of financial time series, it's quintessential to hone in on certain characteristic features:

Trend: Emanating as the backbone of a time series, the trend signifies the overarching movement or trajectory coursing through data over an extended time horizon. Whether it's an ascendant (bullish) arc or a descendent (bearish) curve, the trend offers a panoramic view of an asset or market's health and momentum.

Seasonality: This facet of time series is reminiscent of the cyclical ebb and flow observed at predictable intervals. Be it the festive fervor of holidays, the cyclic cadence of economies, or other recurrent phenomena, seasonality uncovers the rhythmic dance of data and its ties to extrinsic triggers.

Volatility: Serving as the pulse check of a time series, volatility quantifies the magnitude and frequency of data fluctuations. Where heightened volatility speaks of dramatic oscillations and unpredictability, a subdued volatility narrative paints a picture of calm and constancy. In the financial sphere, gauging volatility becomes the linchpin for endeavors ranging from risk stratification to options calibration and portfolio design.

Autocorrelation: Venturing into the realm of self-similarity, autocorrelation deciphers the interplay between a time series and its temporal shadows (lags). It becomes a beacon to discern whether the whispers of the past cast an echo on the present and future. A positive autocorrelation heralds the influence of bygone values on forthcoming ones, while its negative counterpart hints at an antithetical dynamic.

Non-stationarity: Within the tapestry of a time series lies the subtle (or sometimes glaring) shifts in its statistical persona, manifesting as non-stationarity. Such data narratives, devoid of static statistical anchors like mean and variance, often unfurl trends, cyclic patterns, or other time-tethered motifs. Confronting non-stationarity becomes a sine qua non for ensuring precision in analysis and prognostication.

In essence, financial time series is a rich mosaic, intricately woven with these defining traits. A nuanced grasp of elements like trend, seasonality, volatility, autocorrelation, and non-stationarity not only illumines the path for analysts but also paves the way for crafting models that resonate with the symphony of financial data.

## Dependent Variable

$$
 Target(y_t)= 
\begin{cases}
    1,& \text{if } p_{t+1}> 1.004*p_t\\
    0,              & \text{otherwise}
\end{cases}
$$

The target variable, symbolized as $y_t$, can be bifurcated as follows:

The Upward Trend Indicator (1): If the subsequent trading day's closing price, denoted as $p_{t+1}$, surges beyond 1.004 times the present day's closing price, $p_t$, then the target $y_t$ is flagged with a '1'. This indicates the anticipation of a bullish market scenario for the stock.

The Neutral or Downward Trend Indicator (0): If $p_{t+1}$ doesn't scale beyond 1.004 times $p_t$, $y_t$ takes on a value of '0'. This implies an anticipated neutral or bearish market disposition.

In essence, this binary encapsulation of $y_t$ equips our model to discern days radiating bullish vibes from those mirroring a bearish or neutral sentiment.


```python
# Defining the target based on the refined description
combined_features['Target'] = np.where(combined_features['Close'].shift(-1) > 1.004* combined_features['Close'], 1, 0)
combined_features = combined_features.dropna()

# Extracting the feature matrix X
X = combined_features.drop(['Target'], axis=1)

# Pinpointing the target variable y
y =combined_features['Target'].iloc[:,0]
```

## K-Means Clustering

K-Means clustering serves as a powerful method to group similar data points together, effectively condensing the dimensions by recognizing patterns and relationships among various features. This technique operates through several distinct steps, as outlined below:

1. Initialization:

Choose the number of clusters, represented by 'K'.
Randomly initialize 'K' cluster centroids derived from the dataset.

2. Assigning Data Points to Clusters:

For each data point 'xᵢ', compute its distance to each cluster centroid 'cⱼ' using a standard distance metric, often the Euclidean distance.
Assign 'xᵢ' to the cluster whose centroid 'cⱼ' exhibits the smallest distance to 'xᵢ'.

3. Updating Cluster Centroids:

Once all data points are allocated to clusters, calculate the mean of data points within each cluster. This average value becomes the new centroid for the respective cluster.

4. Iterative Process:

Iteratively repeat the steps of assigning data points to clusters and updating cluster centroids until convergence is achieved or a predetermined number of iterations is reached.

5. Termination:

The algorithm concludes when the assignments of data points to clusters no longer significantly change, or upon reaching a specified iteration limit.
The primary objective of K-Means is to identify cluster centroids that minimize the objective function, which is represented mathematically as:

$J = \sum_{i=1}^{n} \sum_{j=1}^{K} ||x_i - c_j||^2$

Here:

'n' signifies the count of data points.

'xᵢ' denotes the ith data point.

'cⱼ' represents the centroid of cluster 'j'.

'K' signifies the number of clusters.

The algorithm aims to discover the optimal cluster centroids that minimize this objective function. However, it's worth noting that K-Means can be sensitive to the initial placement of centroids and might converge to local optima. To address this, it's common practice to run the algorithm multiple times with varying initializations, ultimately selecting the outcome that yields the lowest value for the objective function.

```python
# Import sklearn modules
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
# Define the range of cluster numbers
cluster_range = range(2, 30)

# Calculate inertia for different cluster numbers
cluster_inertia = []
for num_clusters in cluster_range:
    kmeans_model = KMeans(n_clusters=num_clusters)
    kmeans_model.fit(X)  
    cluster_inertia.append(kmeans_model.inertia_)

# Plot relative inertia
plt.plot(cluster_range, np.divide(cluster_inertia, cluster_inertia[0]))
plt.hlines(0.1, cluster_range[0], cluster_range[-1], 'r', linestyles='dashed')
plt.hlines(0.05, cluster_range[0], cluster_range[-1], 'r', linestyles='dashed')
plt.xlabel('Number of Clusters')
plt.ylabel('Relative Inertia')
plt.legend(['Inertia', '10% Relative Inertia', '5% Relative Inertia'])
plt.title('Elbow Method for Optimal Cluster Selection')
plt.show()

# Perform KMeans clustering with optimal cluster number
optimal_num_clusters = 5
kmeans_optimal_model = KMeans(n_clusters=optimal_num_clusters)
kmeans_optimal_model.fit(X)  
optimal_labels = kmeans_optimal_model.predict(X)

# Normalize and fit the model using a pipeline
pipeline = Pipeline([
    ("scaling", MinMaxScaler()),
    ("clustering", KMeans(n_clusters=optimal_num_clusters))
])

# Fit the pipeline model
pipeline.fit(X.T)  # Transpose 'X' to have features as rows

# Assign labels
pipeline_labels = pipeline.predict(X.T)

# Create a DataFrame to hold cluster details
feature_names = X.columns  
cluster_details = pd.DataFrame({
    'Cluster': pipeline_labels,
    'Features': feature_names,
    'Mean_Feature_Value': X.mean(axis=0) 
}).sort_values(by=['Cluster'])

cluster_details.reset_index(drop=True, inplace=True)

# Plot features against clusters
plt.figure(figsize=(20, 10))
plt.scatter(cluster_details.Features, cluster_details.Cluster)
plt.xlabel('Features')
plt.ylabel('Cluster')
plt.title('Features Clustering')
plt.xticks(rotation=90)
plt.show()
```

## Self-Organizing Maps (SOM)

Self-Organizing Maps, a type of artificial neural network, are employed to create a low-dimensional representation of the input data while preserving its topological properties. SOM maps high-dimensional data onto a grid of lower dimensions, which helps capture complex relationships between features. This reduced representation aids in visualization and can serve as an input to subsequent modeling steps.

Total Quantization Error Formula

The total quantization error (E) in a SOM is calculated using the following formula:

$E = \sum_{i=1}^{N} \sum_{j=1}^{M} d_{ij} ||\mathbf{x} - \mathbf{w}_{ij}||^2$

$\begin{align*}
E & : \text{Total quantization error} \\
N, M & : \text{Dimensions of the SOM grid} \\
d_{ij} & : \text{Similarity measure between winning neuron and neuron at position } (i,j) \\
\mathbf{x} & : \text{Input data vector} \\
\mathbf{w}_{ij} & : \text{Weight vector of the neuron at position } (i,j)
\end{align*}$


```python
from sklearn.preprocessing import MinMaxScaler
from minisom import MiniSom

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Scale the feature matrix 'X'
scaled_features = scaler.fit_transform(X)

# Determine the grid size for SOM (thumb rule: grid size > 5 * sqrt(n), considering a 20x20 grid)
grid_size = 20
som = MiniSom(grid_size, grid_size, scaled_features.shape[1], learning_rate=0.5, sigma=5, random_seed=42)

# Initialize SOM with random weights
som.random_weights_init(scaled_features)

# Train the SOM with 10000 iterations
som.train_batch(scaled_features, 10000, verbose=True)

# List the corresponding features for each node on the SOM
node_features = []
for idx in range(len(X)):
    winner_node = som.winner(scaled_features[idx])
    node_features.append((winner_node[0], winner_node[1], X.columns[winner_node[0]], X.columns[winner_node[1]]))

# Create a DataFrame to hold cluster details
df_cluster = pd.DataFrame(node_features, columns=['Node_X', 'Node_Y', 'Feature_X', 'Feature_Y'])

# Plot the SOM visualization
plt.figure(figsize=(20, 10))
for node_info in node_features:
    plt.text(node_info[0], node_info[1], node_info[2], bbox=dict(facecolor='white', alpha=0.5, lw=0))
plt.imshow(som.distance_map())
plt.grid(False)
plt.title('Self Organizing Map Visualization')
plt.show()

# Print the corresponding features for each node
#for node_info in node_features:
#    print(f"Node: ({node_info[0]}, {node_info[1]}), Corresponding Features: {node_info[2]}, {node_info[3]}")
```

## Long Short Term Memory Model

Long Short-Term Memory (LSTM) represents a specialized form of recurrent neural networks (RNNs) engineered to counter the constraints faced by conventional RNNs in grasping long-range correlations within sequential data. Renowned for its prowess in time series forecasting, particularly in financial realms, LSTMs exhibit a remarkable capacity to comprehend intricate temporal associations and circumvent the issue of vanishing gradients.

Theory and Practical Utility:

In financial time series prognosis, the sequential essence of data holds paramount importance, necessitating the apprehension of interdependencies across diverse time spans for precise predictions. Conventional RNNs encounter challenges in handling extensive dependencies and grapple with gradient vanishing concerns, constraining their ability to retain and transmit information across prolonged sequences.

LSTMs navigate these hurdles via specialized components like cells, gates, and memory states. These elements empower LSTMs to grasp and retain information over extensive durations, rendering them apt for tasks encompassing stock price forecasts, volatility estimations, and other financial metrics. Their applications span portfolio optimization, risk management, algorithmic trading, and assorted financial analytical endeavors.

LSTM Architecture:

The LSTM framework revolves around four crucial components: cell state, input gate, forget gate, and output gate.

Cell State: Functions as an enduring memory unit retaining both short-term and long-term data, subject to selective alterations via gating mechanisms.

Input Gate: Dictates updates to the cell state by evaluating input data and previous hidden states, employing sigmoid activation for selective information update.

Forget Gate: Determines information retention or discard from the cell state, akin to the input gate but deciding what to remember or forget.

Output Gate: Controls the output from the cell state, processing current input and prior hidden states to produce the final output.

Input Gate:

$
\begin{align*}
i_t & = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i) \\
\tilde{C}_t & = \text{tanh}(W_c \cdot [h_{t-1}, x_t] + b_c) \\
C_t & = f_t \cdot C_{t-1} + i_t \cdot \tilde{C}_t
\end{align*}
$

Forget Gate:

$
f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)
$

Output Gate:

$
\begin{align*}
o_t & = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o) \\
h_t & = o_t \cdot \text{tanh}(C_t)
\end{align*}
$

Where:

- $(x_t)$ = Input data at time \(t\),
- $(h_{t-1})$ = Previous hidden state,
- $(b_i)$, $(b_f)$, $(b_o)$, $(b_c)$ = Bias terms,
- $(\sigma)$ = Sigmoid activation function,
- $(\text{tanh})$ = Hyperbolic tangent activation function.




Conclusion:
LSTMs excel in capturing temporal intricacies, leveraging historical patterns to make informed predictions. Their prowess in handling financial time series data, where past insights dictate future trends, positions them as indispensable tools for traders, analysts, and researchers in the financial sphere.

```python
# Function to generate sequential data sequences and targets
def generate_sequences(data, target, sequence_length):
    sequences = []
    targets = []
    
    # Generate sequences and corresponding targets
    for i in range(len(data) - sequence_length):
        sequences.append(data[i:i+sequence_length])
        targets.append(target[i+sequence_length])
    
    return np.array(sequences), np.array(targets)

# Define the sequence length for creating sequences
sequence_length = 30  

# Generate sequences and targets for training and testing sets
X_train_seq, y_train_seq = generate_sequences(X_train, y_train, sequence_length)
X_test_seq, y_test_seq = generate_sequences(X_test, y_test, sequence_length)
y_train_seq=y_train_seq
y_test_seq=y_test_seq
```

## Bidirectional LSTM

Bidirectional LSTM, a variant of the traditional LSTM, significantly enhances its comprehension of sequential data by processing information in two directions: both forward and backward through time steps. Unlike the standard LSTM that solely considers past inputs for predictions, the bidirectional architecture incorporates information from both past and future time steps.

Standard LSTMs focus solely on historical data to understand patterns, lacking the foresight to leverage future information. Bidirectional LSTMs overcome this by employing two distinct sets of hidden states for processing sequences:

Forward Pass: Sequential data is processed from the sequence's start to its end. At each time step, the forward LSTM cell updates its hidden state by considering the current input alongside the previous hidden state. This captures historical context and patterns within the sequence.

Backward Pass: In contrast, the sequence is also processed in reverse, from the end to the start. Similar to the forward pass, the backward LSTM cell updates its hidden state based on the current input and the previous hidden state, capturing potential future dependencies.

Information Fusion: Outputs from both passes are merged at each time step through concatenation. This fusion combines insights gathered from past and future directions, creating a comprehensive representation that enhances predictions or further data processing.

```python
# Define the number of time steps and features
time_steps = X_train_seq.shape[1]  # Number of time steps in sequences
num_features = X_train_seq.shape[2]  # Number of features in each time step

# Build the bidirectional LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(units=100, return_sequences=True), input_shape=(time_steps, num_features)))
model.add(Bidirectional(LSTM(units=100, return_sequences=True)))
model.add(Bidirectional(LSTM(units=100)))
model.add(Dense(units=1, activation='sigmoid'))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_seq, y_train_seq, epochs=100, batch_size=32, validation_data=(X_test_seq, y_test_seq),verbose=0)

# Evaluate the model
loss, accuracy = model.evaluate(X_test_seq, y_test_seq)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")
```

Hyperparameter tuning, at its core, is about finding that sweet spot – the right combination of settings that maximize the model's performance. It's like trying out various dials and knobs on a machine to get the best output. With techniques like Grid Search, we systematically explore different setups by testing the model with various combinations of these settings. It's like methodically trying out different ingredient proportions in a recipe to find the perfect balance for taste. Ultimately, the goal is to identify the settings that make the model most accurate, stable, and efficient.

```python
# Define the hyperparameter search space
def build_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32), 
                   activation='relu', input_shape=(sequence_length, X_train_seq.shape[2])))
    model.add(Dense(units=1))
    model.compile(optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='mse')
    return model

# Initialize the tuner
tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,  # Number of trials
    executions_per_trial=1,
    directory='tuner_results'  # Directory to store results
)

# Search for the best hyperparameters
tuner.search(X_train_seq, y_train_seq, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Get the best model and summary of the search
best_model = tuner.get_best_models(num_models=1)[0]
tuner.results_summary()

# Evaluate the best model on test data
test_loss = best_model.evaluate(X_test_seq, y_test_seq, verbose=0)
print(f"Test Loss for Best Model: {test_loss:.4f}")
```

## Understanding ROC & AUC in Classification

The ROC (Receiver Operating Characteristic) curve acts as a pivotal evaluation tool, especially in binary classification scenarios. It displays the trade-off between two essential metrics: TPR (True Positive Rate) and FPR (False Positive Rate), each determined by varying classification thresholds.

TPR, also called sensitivity or recall, highlights the proportion of accurately predicted positive instances out of all actual positives. Essentially, it measures how well the model identifies positive cases correctly.

In contrast, FPR represents the proportion of incorrectly predicted positive instances out of all actual negatives. It reveals instances where the model misclassifies negatives as positives.

The ROC curve visually represents these metrics across different thresholds. Each point on the curve corresponds to a specific threshold setting. Analyzing the ROC curve provides insights into the model's ability to balance TPR and FPR across diverse thresholds. The ideal scenario is a curve closer to the top-left corner, signifying better overall performance.

AUC (Area Under the Curve) is a comprehensive metric that quantifies the model's performance. It computes the area under the ROC curve, presenting a value between 0 and 1. A higher AUC suggests superior discrimination between positive and negative instances. An AUC of 0.5 represents random guessing, while an AUC of 1 indicates flawless discrimination.

```python
# Calculate AUC
roc_auc = roc_auc_score(y_test_seq, y_pred_prob)
print(f"AUC: {roc_auc:.4f}")

# Plot ROC Curve
fpr, tpr, _ = roc_curve(y_test_seq, y_pred_prob)
roc_auc_curve = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_curve)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

## Confusion Matrix in Binary Classification

The confusion matrix is the cornerstone of assessing a binary classification model's performance. It's akin to a tabular representation that showcases the model's predictions against the actual labels. This square matrix comprises four crucial components:

True Positives (TP): Instances where the model correctly predicts the positive class. In stock price prediction, this could mean accurately identifying upward trends in Fidelity National Information Services (FIS) stock prices.

True Negatives (TN): Instances where the model correctly predicts the negative class. For stock prices, this indicates the correct anticipation of downward trends or the absence of positive movements.

False Positives (FP): Incorrect predictions of the positive class when the actual label is negative. These "false alarms" denote wrongly predicted upward trends.

False Negatives (FN): Incorrect predictions of the negative class when the actual label is positive. Representing "misses," these instances signify a failure to predict positive trends.
